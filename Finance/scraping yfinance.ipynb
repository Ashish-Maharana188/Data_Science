```python
%matplotlib inline

import matplotlib
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Set up yfinance
yf.pdr_override()

# Create a Ticker object for GME stock
gme = yf.Ticker("GME")

# Retrieve historical data for the stock
gme_hist = gme.history(period="max")

# Reset the index of the DataFrame
gme_hist_reset = gme_hist.reset_index()

# Convert datetime values to timezone-naive
gme_hist_reset['Date'] = gme_hist_reset['Date'].dt.tz_localize(None)

# Plot GameStop stock graph
def make_graph(data_frame, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data_frame['Date'], data_frame['Close'], label='Close Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Call the function to plot the GameStop stock graph
make_graph(gme_hist_reset, "GameStop Stock Data")
```
